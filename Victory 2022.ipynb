```python
#Programs a simulation of our BLAST draft with keepers
#Uses lists of lists as storage, and dataframes to print in organized fashion only
import pandas as pd
import colorama
from colorama import Fore, Back, Style
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import csv


#Define rankings as a list of lists - each list is simply the name and the position
# the index of rankings will serve to assist us a lot
colorama.init(autoreset=True)#auto resets your settings after every output

#rankings = [['Jonathan Taylor', 'RB'], ['Josh Allen', 'QB'], ['Patrick Maholmes', 'QB'], ['Cooper Kupp', 'WR'], ['Derrick Henry', 'RB'], ['Austin Ekeler', 'RB'],['Najee Harris', 'RB'], ['Alvin Kamara', 'RB'], ['Davante Adams', 'WR'], ['Dalvin Cook', 'RB'], ['Deebo Samuel', 'WR'], ['Joe Mixon', 'RB'], ['Christian McCaffrey', 'RB'], ['Justin Jefferson', 'WR'], ['Jamaar Chase', 'WR'], ['Justin Herbert', 'QB'], ['Deandre Swift', 'RB'], ['Nick Chubb', 'RB'], ['James Conner', 'RB'], ['Joe Burrow', 'QB'], ['Stefon Diggs', 'WR'], ['Tyreek Hill', 'WR'],  ['George Kittle', 'TE'], ['AJ Brown', 'WR'], ['Leonard Fournette', 'RB'], ['Javonte Williams', 'RB'], ['Jaylen Waddle', 'WR'], ['Keenan Allen', 'WR'], ['Lamar Jackson', 'QB'], ['Ceedee Lamb', 'WR'], ['Mark Andrews', 'TE'], ['Aaron Jones', 'RB'], ['Deandre Hopkins', 'WR'],  ['Mike Evans', 'WR'], ['David Montgomery', 'RB'], ['Cam Akers', 'RB'], ['JK Dobbins', 'RB'], ['Tee Higgins', 'WR'], ['Kyle Pitts', 'TE'], ['Aaron Rodgers', 'QB'], ['Travis Kelce', 'TE'], ['Antonio Gibson', 'RB'],  ['Matthew Stafford', 'QB'],  ['Kyler Murray', 'QB'], ['Marquise Brown', 'WR'], ['Trey Lance', 'QB'], ['Damien Harris', 'RB'], ['Darren Waller', 'TE'], ['Chris Godwin', 'WR'], ['Brandin Cooks', 'WR'], ['Tom Brady', 'QB'], ['Clyde Edwards-Helaire', 'RB'], ['Dak Prescott', 'QB'], ['Jalen Hurts', 'QB'], ['Eli Mitchell', 'RB'],  ['Robert Woods', 'WR'], ['Terry McLaurin', 'WR'], ['Ezekiel Elliott', 'RB'], ['Saquon Barkley', 'RB'], ['DJ Moore', 'WR'], ['Chase Edmonds', 'RB'], ['Darnell Mooney', 'WR'], ['Michael Thomas', 'WR'], ['Travis Etienne', 'RB'], ['Amon Ra St. Brown', 'WR'], ['Russell Wilson', 'QB'], ['Dawson Knox', 'TE'], ['Tyler Boyd', 'WR'], ['Devonta Smith', 'WR'], ['Diontae Johnson', 'WR'], ['Kirk Cousins', 'QB'], ['Ryan Tannehill', 'QB'], ['Josh Jacobs', 'RB'], ['Brandon Aiyuk', 'WR'], ['Devin Singleterry', 'RB'], ['DK Metcalf', 'WR'], ['AJ Dillon', 'RB'],  ['Miles Sanders', 'RB'], ['Tua Tagovailoa', 'QB'], ['Mac Jones', 'QB'], ['Hunter Renfrow', 'WR'], ['TJ Hockenson', 'TE'], ['Zach Ertz', 'TE'], ['Michael Carter', 'RB'], ['Elijah Moore', 'WR'], ['Chris Carson', 'RB'], ['Deshaun Watson', 'QB'], ['Kareem Hunt', 'RB'], ['Odell Beckham Jr.', 'WR'], ['Amari Cooper', 'WR'], ['Michael Pittman', 'WR'], ['Justin Fields', 'QB'],  ['Dallas Goedert', 'TE'], ['Pat Freiermuth', 'TE'], ['Hunter Henry', 'TE'], ['Kadarius Toney', 'WR'], ['Dalton Shultz', 'TE'], ['Devante Parker', 'WR'], ['Mike Gesicki', 'TE'], ['Mike Williams', 'WR'], ['Gabriel Davis','WR'], ['Braxton Berrios', 'WR'],  ['Allen Robinson', 'WR'], ['Sony Michel', 'RB'], ['Derek Carr', 'QB'], ['Melvin Gordon', 'RB'], ['Trevor Lawrence', 'QB'], ['Rhamondre Stevenson', 'RB'], ['Tony Pollard', 'RB'], ['Rashaad Penny', 'RB'], ['Davis Mills', 'QB'], ['Jerry Jeudy', 'WR'], ['James Robinson', 'RB'], ['Carson Wentz', 'QB'], ['Juju Smith-Schuster', 'WR'], ['Jimmy Garropolo', 'QB'], ['Alexander Mattison', 'RB'], ['Xade Jones', 'WR'], ['Courtland Sutton', 'WR'], ['Van Jefferson', 'WR'], ['Curtis Samuel', 'WR'], ['Laviska Shenault', 'WR'], ['Jared Goff', 'QB'], ['Darrell Williams', 'RB'], ['Christian Kirk', 'WR'], ['Jakobi Meyers', 'WR'], ['Chase Clayool', 'WR'], ['Josh Palmer', 'WR'], ['Rondale Moore', 'WR'], ['Rashod Bateman', 'WR'], ['Isaiah Mackenzie', 'WR'], ['Tyler Lockett', 'WR'], ['Ronald Jones', 'RB']]    
columns = ['player', 'position']

with open('player_output.csv', 'r') as f: 
    reader = csv.reader(f) 
    next(reader, None)  # skip the headers
    rankings = list(reader)

stoosh = [['Deebo Samuel', 'WR'], ['Ezekiel Elliott', 'RB'], ['Chris Godwin', 'WR'], ['David Montgomery', 'RB'], ['Saquon Barkley', 'RB']]
silentmajority = [['Justin Jefferson', 'WR'], ['Jamaar Chase', 'WR'], ['Lamar Jackson', 'QB'], ['Stefon Diggs', 'WR'], ['Nick Chubb', 'RB'], ['Kyle Pitts', 'TE']]
homohominilupus = [['Deandre Hopkins', 'WR'], ['Diontae Johnson', 'WR'], ['Tyler Lockett', 'WR'], ['Patrick Maholmes', 'QB'], ['Najee Harris', 'RB'], ['Joe Burrow', 'QB'], ['Travis Kelce', 'TE']]
thefightingpaulannes = [ ['DK Metcalf', 'WR'],  ['Aaron Rodgers', 'QB'], ['Aaron Jones', 'RB'], ['Christian McCaffrey', 'RB'], ['Keenan Allen', 'WR'], ['Russell Wilson', 'QB'], ['Deandre Swift', 'RB'], ['Leonard Fournette', 'RB']]
lettherebelight = [['Travis Etienne', 'RB'], ['Trey Lance', 'QB'], ['Michael Thomas', 'WR'], ['Terry McLaurin', 'WR'], ['Kyler Murray', 'QB'], ['Dalvin Cook', 'RB'], ['AJ Brown', 'WR'], ['Javonte Williams', 'RB'], ['Jaylen Waddle', 'WR'], ['Ceedee Lamb', 'WR'], ['Cam Akers', 'RB']]
cambodiancojones = [['Davante Adams', 'WR'], ['Alvin Kamara', 'RB'], ['George Kittle', 'TE'], ['Mike Evans', 'WR'], ['AJ Dillon', 'RB']]
billkates = [['Derrick Henry', 'RB'], ['James Conner', 'RB'], ['Josh Allen', 'QB'], ['Austin Ekeler', 'RB'], ['Mark Andrews', 'TE'], ['Clyde Edwards-Helaire', 'RB'],  ['Cooper Kupp', 'WR'], ['Dak Prescott', 'QB']]
olddog = [['Jonathan Taylor', 'RB'], ['Tyreek Hill', 'WR'], ['Joe Mixon', 'RB'], ['Justin Herbert', 'QB'], ['Tee Higgins', 'WR'], ['Eli Mitchell', 'RB'], ['JK Dobbins', 'RB'], ['Amon Ra St. Brown', 'WR'], ['Matthew Stafford', 'QB']] 

teams = [stoosh, silentmajority, homohominilupus, thefightingpaulannes, lettherebelight, cambodiancojones, billkates, olddog]

stoosh2 = []
silentmajority2 = []
homohominilupus2 = []
thefightingpaulannes2 = []
lettherebelight2 = []
cambodiancojones2 = []
billkates2 = []
olddog2 = []

team_names = ['Stoosh', 'Silent Majority', 'Homo Homini Lupus', 'The Fighting Paulannes', 'Let There Be Light', 'Cambodian Cojones', 'Bill Kates', 'Old Dog']
new_teams = [stoosh2, silentmajority2, homohominilupus2, thefightingpaulannes2, lettherebelight2, cambodiancojones2, billkates2, olddog2]

available = rankings.copy()

#prints rankings' top 25 as a dataframe
def print_rankings(players):
    index = 0
    df = pd.DataFrame(players, columns = ['Name', 'Position'])
    df.index += 1
    print(df.loc[:25])
    ''''
    for each in players:
        index += 1
        if index < 21: print(Back.BLACK + Fore.CYAN + str(index) + str(each))#Revises the list of available players based on rankings and those that have been selected having already been removed.  
    '''
    
#Returns sorted list of available (sorted by rankings)
def sort_available():
    new_available = []
    for player in available:
        new_available.insert(rankings.index(player), player)
    return new_available

#Prints according to position from whatever players it receives, a position argument and number of players to group
def print_grouped(players, pos, num):
    df = pd.DataFrame(players, columns = ['Name','Position'])
    df.index += 1
    df_selected = df.loc[df.Position == pos, :].iloc[:num]
    print(df_selected)
    players_lst = df_selected.values.tolist()
    return players_lst 

qbstrength = 0
rbstrength = 0
wrstrength = 0
testrength = 0

# Takes rankings of current top 5 from each position group (received as df arg) and assigns them a strength score
# Strength depends on lowness of their rankings.  Each ranking is subtracted from 100 to create a score.  
#Scores are then added to the strength score and returned.
def score_top5(players):
    group_score = 0
    for player in players:
        #print(player)
        score = 99 - rankings.index(player)
        group_score += score
    return group_score

# Prints bar chart showing the relative strengths at the top 5 of each position
# Receives a dictionary argument with strengths by position (keys)
def print_bar_chart(all_strengths, title):
    plt.bar(all_strengths.keys(), all_strengths.values(), color=['red', 'green', 'blue', 'cyan'])
    plt.yscale('log')
    plt.ylabel('Score')
    plt.title(title)
    #plt.gca().yaxis.set_minor_formatter(mticker.PercentFormatter())
    #plt.gca().ticklabel_format(useOffset=False, style='plain')
    plt.show()
    
#Prints rankings' top 25 and then grouped top 5 per position, then bar chart showing position strength for available
def print_ranks_grouped():
    available_current = sort_available()
    print_rankings(available_current)
    qbs = print_grouped(available_current, 'QB', 5)
    rbs = print_grouped(available_current, 'RB', 5)
    wrs = print_grouped(available_current, 'WR', 5)
    tes = print_grouped(available_current, 'TE', 5)
    
    qbstrength = score_top5(qbs)
    rbstrength = score_top5(rbs)
    wrstrength = score_top5(wrs)
    testrength = score_top5(tes)
    all_strengths = {'QB Strength': qbstrength, 'RB Strength':rbstrength, 'WR Strength': wrstrength, 'TE Strength': testrength}
    
    print_bar_chart(all_strengths, 'Available')
    print('')

# Calculates strength of team and prints bar chart showing strength of team
# Receives team as argument, list of lists
def strength_of_team(team, name):
    qbs = print_grouped(team, 'QB', 3)
    rbs = print_grouped(team, 'RB', 3)
    wrs = print_grouped(team, 'WR', 3)
    tes = print_grouped(team, 'TE', 3)
    
    qbstrength = score_top5(qbs)
    rbstrength = score_top5(rbs)
    wrstrength = score_top5(wrs)
    testrength = score_top5(tes)
    all_strengths = {'QB Strength': qbstrength, 'RB Strength':rbstrength, 'WR Strength': wrstrength, 'TE Strength': testrength}
    
    print_bar_chart(all_strengths, name)
    print('')
    
def getranking(player_name, pos):
    index = 0
    for each in rankings:
        index += 1
        if player_name == each[0] and pos == each[1]:
            return index

#should rerank players based on receiving old rank and new rank only
def rerank(old_rank, new_rank):
    player_to_add = rankings.pop(old_rank-1)
    print("You are moving", player_to_add)
    rankings.insert(new_rank-1, player_to_add)
    available = rankings.copy()

#returns index of lowest rank on the team it receives
def lowest_rank(team):
    lst_rank = 0
    for player in team:
        if lst_rank < rankings.index(player):
            lst_rank = rankings.index(player)
    return lst_rank

#Returns lowest ranking player on the team it receives
def lowest_ranking_player(team):
    for player in team:
        if lowest_rank(team) == rankings.index(player):
            return rankings.index(player)

#if the player is not already on the team, it removes the player from team and inserts player to the front of the list
#then returns the changed team
def replace_lowest(team, player):
    if player not in team:
        team.remove(rankings[lowest_ranking_player(team)])
        team.insert(0, player)
    return team

#define a function that determines what each team's first three picks are based on the players from their teams with the top 3 ranking
def select_first_three():
    #Adds selected players according to rankings and then prints teams
    for t in range(len(teams)):
        for p in range(len(teams[t])):
            player = teams[t][p]
            if len(new_teams[t]) < 3:
                new_teams[t].append(player)
                #print('just added player', player)
            if len(new_teams[t]) == 3 and rankings.index(player) < lowest_rank(new_teams[t]):
                #print("Lowest rank of team ", new_teams[t], "is ", lowest_rank(new_teams[t])) 
                if rankings[lowest_rank(new_teams[t])] != player:
                    #print("Replacing", rankings[lowest_rank(new_teams[t])], " with ", player)
                    new_teams[t] = replace_lowest(new_teams[t], player)
        print(team_names[t])
        print(pd.DataFrame(new_teams[t], columns = ['Name', 'Position']).to_string(index = False))
        print("")
    
    #Removes selected players from available list
    for t in range(len(teams)):
        for p in range(len(teams[t])):
            player = teams[t][p]
            if player in new_teams[t] and player in available:
                available.remove(player)

def send_to_rerank():
    print("So you'd like to rerank some players on the list first? Y/N")
    rerank_bool = input()
    while rerank_bool == 'Y':
        print("First, enter the current number next to the player you'd like to rerank:")
        old_rank = input()
        print("Ok. Now enter the new number at which you'd like to rank that player:")
        new_rank = input()
        rerank(int(old_rank), int(new_rank))
        print_ranks_grouped()
        print("Now, do you want to continue to rerank? Y/N")
        rerank_bool = input()
    

print(Back.YELLOW + "Would you like to see the rankings?(Y/N/R - R for re-rank)")
x = input()
if x == 'Y':
    print_ranks_grouped()  
elif x == 'R': send_to_rerank()


print("Would you like to have the teams select keepers?(Y/N/R - R for re-rank)")
first_rounds = input()
if first_rounds == 'Y':
    print(Fore.RED + Back.YELLOW + "This is the composition of each team after keepers have been selected:")
    select_first_three()
    print(Fore.RED + Back.YELLOW + "And the following players are still available:")
    print_ranks_grouped()
else:
    send_to_rerank()
    print(Fore.RED + Back.YELLOW + "This is the composition of each team after keepers have been selected:")
    select_first_three()
    print(Fore.RED + Back.YELLOW + "And the following players are still available:")
    print_ranks_grouped()
    
    
        
        
```

```python
# Hand-Select Simulation
# This cell will allow me to hand-pick the next player for each team until each team is full

#returns true if roster is filled up
def filled_up(team):
    #count players at each position
    qbs = 0
    wrs = 0
    rbs = 0
    tes = 0
    qbs_full = False
    wrs_full = False
    rbs_full = False
    tes_full = False
    for player in team:
        if player[1] == 'QB':
            qbs += 1
        if player[1] == 'WR':
            wrs += 1
        if player[1] == 'RB':
            rbs += 1
        if player[1] == 'TE':
            tes += 1
    if qbs + rbs + wrs + tes >= 15:
        return True
    else: return False

def all_filled_up():
    all_filled = True
    for team in new_teams:
        if not filled_up(team):
            all_filled = False
            return False
    return all_filled

def input_pick(team):
    print("Team: ", team_names[new_teams.index(team)])
    print("Please input next pick using number (R for re-rank, S for team strength)")
    next_pick = input()
    print('------------------------------------END OF PICK-----------------------------')
    if next_pick == 'R' or next_pick == 'r': 
        send_to_rerank()
        print("Please input next pick using number (R for re-rank, S for team strength)")
        next_pick = input()
        team.append(available[int(next_pick)-1])
        available.remove(available[int(next_pick)-1])
    elif next_pick == 'S' or next_pick == 's':
        strength_of_team(team, team_names[new_teams.index(team)])
        print("Please input next pick using number (R for re-rank, S for team strength)")
        next_pick = input()
        team.append(available[int(next_pick)-1])
        available.remove(available[int(next_pick)-1])
    else: 
        team.append(available[int(next_pick)-1])
        available.remove(available[int(next_pick)-1])

level = 4

while not all_filled_up():
    for team in new_teams:
        print("Round ", level, "Available:")
        print_ranks_grouped()
        if not filled_up(team) and level % 2 == 0:
            print(pd.DataFrame(team, columns = ['Name', 'Position']).to_string(index = False))
            input_pick(team)
        else: break
    level += 1
    for team in reversed(new_teams):
        print("Round 1", level, "Available:")
        print_ranks_grouped()
        if not filled_up(team) and level % 2 == 1:
            print(pd.DataFrame(team, columns = ['Name', 'Position']).to_string(index = False))
            input_pick(team)
        else: break
    level += 1
        
```
